```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
def Gradient(x,y):
    m = 0
    c = 0
    #take iterations and learning rate, keep revising it
    iterations = 1000
    learning_rate = 0.001
    n = len(x)
    
    for i in range(iterations):
        gap = y - (m*x+c)
        
        #error
        error = 1/n * sum(val**2 for val in (gap))
        
        dm = (-2/n) * sum(x*gap)
        dc = (-2/n) * sum(gap)
        
        #change the m and c
        m = m - learning_rate * dm
        c = c - learning_rate * dc
        
        print("iteration{} Error {} m {} c {}".format(iterations, m, c, error))
    
x = np.array([5,7,8,9,11])
y = np.array([10,12,15,18,19])
             
Gradient(x,y)
```
